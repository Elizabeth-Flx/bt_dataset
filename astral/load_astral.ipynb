```python
# Filter out contaminants and decoys

df_merged = df_merged[df_merged['Reverse'].isna()]
df_merged = df_merged[df_merged['Potential contaminant'].isna()]

# Filter out sequences with length > 30

df_merged = df_merged[df_merged['Length'] <= 30]

df_merged
```

```python
map_keys = [key for i in range(1,30) for key in 
            ('y%s' % i, 'y%s(2+)' % i, 'y%s(3+)' % i,
             'b%s' % i, 'b%s(2+)' % i, 'b%s(3+)' % i)]
map_values = list(range(174))
fragment_map = dict(zip(map_keys, map_values))

def generate_intensity_annotations(matches, intensities, sequence_len, charge):

    annotation_list = [0] * (sequence_len-1)*6 + [-1] * (29-sequence_len+1)*6
    
    fragment_list = matches.split(';')
    intensity_list = [np.float64(intensity) for intensity in intensities.split(';')]

    fragment_intensities = dict(zip(fragment_list, intensity_list))
    fragment_intensities = {frag: inte for frag, inte in fragment_intensities.items() if frag in map_keys}

    max_intensity = max(fragment_intensities.values())
    
    for frag, inte in fragment_intensities.items():
        annotation_list[fragment_map[frag]] = inte / max_intensity
    
    if charge <= 2:
        for i in range(2,174,3):
            annotation_list[i] = -1
    return np.array(annotation_list, dtype=np.float32)




# i=105
# ex_match = df_merged.iloc[i]['Matches']
# ex_inten = df_merged.iloc[i]['Intensities']
# ex_lengt = df_merged.iloc[i]['Length']
# ex_charg = df_merged.iloc[i]['Charge']

# print(ex_match)
# print(ex_inten)
# print(ex_lengt)
# print(ex_charg)

# print(generate_intensity_annotations(ex_match, ex_inten, ex_lengt, ex_charg))

```

```python
df_merged['intensities_raw'] = df_merged.apply(lambda row: generate_intensity_annotations(row['Matches'], 
                                                                                          row['Intensities'], 
                                                                                          row['Length'], 
                                                                                          row['Charge']), axis=1)

df_merged.reset_index(drop=True, inplace=True)
df_merged
```
