```python
import pandas as pd
import numpy as np
import os
```

```python
filenames = [m for m in os.listdir("../../parquets")]
filenames
```

```python
def reformat_data(filename):

    df = pd.read_parquet(f'../../parquets/{filename}')

    df['prosit_sequence'] = df['modified_sequence'].str.replace('M[UNIMOD:35]', 'm', regex=False).str.replace('[]-', '', regex=False).str.replace('-[]', '', regex=False)
    df = df[~df['prosit_sequence'].str.contains('\\[UNIMOD')]

    df['charge'] = df.apply(lambda row: np.argmax(row['precursor_charge_onehot'])+1, axis=1)
    df['collision_energy'] = df['collision_energy_aligned_normed']
    df['method_nr'] = df.apply(lambda row: 'CID' if row['method_nbr'] == 1 else 'HCD' , axis=1)
    df['machine'] = 'Orbitrap_Fusion_Lumos'

    df.to_parquet(f'./formatted_{filename}')
```

```python
reformat_data(filenames[0])
```

```python
for filename in filenames[1:]:
    print("Formatting", filename)
    reformat_data(filename)
    print("Done", filename)
```

```python
# combine all parquets into one dataframe
df = pd.concat([pd.read_parquet("../../parquets/" + f) for f in filenames])
df
```
