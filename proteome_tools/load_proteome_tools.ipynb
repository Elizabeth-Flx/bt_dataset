```python
import pandas as pd
from utils import tokenizer
import os

filenames = [m for m in os.listdir("parquets")]

dic = {}
for filename in filenames:
    print(filename)

    df = pd.read_parquet('parquets/%s'%filename)

    for modseq in df['modified_sequence']:
        tokseq = tokenizer(modseq)
        
        for token in tokseq:
            if token not in dic:
                dic[token] = 1
            else:
                dic[token] += 1

with open("token_counts.txt", 'w') as f:
    f.write("\n".join(["%s %d"%(m,n) for m,n in dict(sorted(dic.items())).items()]))
```
