```python
import pandas as pd
import h5py

from pyarrow.parquet import ParquetFile
import pyarrow as pa 

import numpy as np

```

```python
hf = "prediction_tof_train.hdf5"

with h5py.File(hf, "r") as f:

    print(f.keys())
    

    hdf5_dict = {}
    
    for key in f.keys():
        hdf5_dict[key] = np.array(f[key]).tolist()

df = pd.DataFrame.from_dict(hdf5_dict)
```

```python
map_alphabet = {
    1: 'A',
    2: 'C',     # fixed modification
    3: 'D',
    4: 'E',
    5: 'F',
    6: 'G',
    7: 'H',
    8: 'I',
    9: 'K',
    10: 'L',
    11: 'M',
    12: 'N',
    13: 'P',
    14: 'Q',
    15: 'R',
    16: 'S',
    17: 'T',
    18: 'V',
    19: 'W',
    20: 'Y',
    21: 'm',    # unfixed modification (oxidation)
    0: ''
}

def one_hot_to_sequence(sequence_hot):
    sequence = ''
    for i in sequence_hot:
        sequence += map_alphabet[i]
    return sequence
```

```python
df['prosit_sequence'] = df.apply(lambda row: one_hot_to_sequence(row['obs_sequence_integer']), axis=1)
df
```
